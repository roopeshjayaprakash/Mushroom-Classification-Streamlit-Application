### Visualisation

```python
fig = plt.figure(figsize=(12, 40))

# Loop through each column in the DataFrame to create a pie chart
for i, col in enumerate(df.columns):
    classlabels = list(df[col].unique())
    percentage = df[col].value_counts(normalize=True) * 100

    ax = plt.subplot(13, 2, i + 1)
    ax.pie(percentage, labels=classlabels, autopct='%1.1f%%', colors=sns.color_palette('pastel', len(classlabels)))
    ax.set_title(f'Pie chart for {col}')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
```

the important point to see is that in the target class poisonous and edible mushrooms are approximately equally distibuted. So the model is not biased.
